```python
import cv2
from matplotlib.pyplot import imshow
import numpy as np
```

```python
naruto_image = cv2.imread("Data/Naruto.png",  cv2.IMREAD_COLOR)
dbz_image = cv2.imread("Data/DBZ.png",  cv2.IMREAD_COLOR)

dbz_new_shape = (naruto_image.shape[1], naruto_image.shape[0]) 
dbz_image = cv2.resize(dbz_image, dbz_new_shape, interpolation = cv2.INTER_CUBIC)
```

```python
imshow(cv2.cvtColor(naruto_image, cv2.COLOR_BGR2RGB) )
```

```python
imshow(cv2.cvtColor(dbz_image, cv2.COLOR_BGR2RGB) )
```

```python
normal_addition = naruto_image + dbz_image
imshow(cv2.cvtColor(normal_addition, cv2.COLOR_BGR2RGB) )
```

```python
no_weight_add = cv2.add(naruto_image, dbz_image)
imshow(cv2.cvtColor(no_weight_add, cv2.COLOR_BGR2RGB) )
```

```python
weight_add = cv2.addWeighted(naruto_image, 0.3, dbz_image, .7, 0)
imshow(cv2.cvtColor(weight_add, cv2.COLOR_BGR2RGB) )
```
