```python
import cv2
from matplotlib.pyplot import imshow
import numpy as np
```

```python
goku_image = cv2.imread("Data/Goku.jpg", cv2.IMREAD_COLOR)
imshow(cv2.cvtColor(goku_image, cv2.COLOR_BGR2RGB) )
```

# Sobel Opeator 
It is the differentiation in one direction ie, either along X or along Y.

```python
sobelx_edge_goku = cv2.Sobel(goku_image, cv2.CV_16U, 1, 0, ksize=3)
imshow(sobelx_edge_goku.astype(np.uint8))
```

```python
sobelx_edge_goku = cv2.Sobel(goku_image, cv2.CV_16U, 0, 1, ksize=3)
imshow(sobelx_edge_goku.astype(np.uint8))
```

# Laplacian
It performs Laplacian operation at each pixel.

```python
laplacian_edge_goku = cv2.Laplacian(goku_image, cv2.CV_16U)
imshow(laplacian_edge_goku.astype(np.uint8))
```

# Canny
- If the position of a body suddenly changes, not in smooth manner (this is what happens when a body hits a wall), then it can be said that the body recieved an impulse. In simmilar way if the intensity of pixel suddenly changes then it can be said as edge. So, if the gradient have high magnitude then it can be said to be edge in image.  
- Canny is a smart algorithm. It first removes the noise using gaussian fiter and then takes the gradient at each pixel. The magnitude of the gradient is taken and the direction of the gradient is rounded to four angle, i.e vertical, horizontal and the two digonal. 
- At each point it rechecks if along the direction of the gradinent, it is a point of local maxima.
- It the discards the edge pixel below certain value and keeps the pixel above another certain value. For the pixel lying between those two values, if the pixel has a neighbouring pixel which is an edge then it is considered to be an edge and if not then it is not considered to be an edge.

```python
canny_edge_goku = cv2.Canny(goku_image, 100, 200)
imshow(canny_edge_goku, cmap='gray')
```
