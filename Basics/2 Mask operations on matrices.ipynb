```python
import cv2
from matplotlib.pyplot import imshow
import numpy as np
```

```python
image = cv2.imread("Data/Taylor.jpg",  cv2.IMREAD_COLOR) 
```

```python
#Adding Sharpness To The Images
kernel = np.array([[0, -1, 0], [-1, 5, -1],
                   [0, -1, 0]], np.float32)

#To sharpen the image, this 3x3 kernal is used 
#for computation over a pixel neighborhood, to 
#obtain the value of the pixel at the given position.
#Here the bit depth is -1, which indicates the bit
#depth same as original image is used.
dst1 = cv2.filter2D(image, -1, kernel)
```

```python
imshow(cv2.cvtColor(dst1, cv2.COLOR_BGR2RGB) )
```
